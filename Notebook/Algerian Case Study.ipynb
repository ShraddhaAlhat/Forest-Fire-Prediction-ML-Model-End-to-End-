```python
X=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
X_train.corr()
```

```python
#multicollinearity
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
                
```

```python
corr_feature=correlation(X_train,0.85) # threshold value is fixed by domain expert
```

```python
corr_feature
```

```python
X_train.drop(corr_feature,axis=1,inplace=True)
X_test.drop(corr_feature,axis=1,inplace=True)

```

```python
X_train.shape,X_test.shape
```

```python
#feature scalling and standardization
from sklearn.preprocessing import StandardScaler
```

```python
Scaler=StandardScaler()
```

```python
X_train_scaled=Scaler.fit_transform(X_train)
```

```python
X_test_scaled=Scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
#Box plot to understand the effect of standardscaler
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_train befor Scalling')
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train after Scalling')
plt.tight_layout()

plt.show()
```

```python
#linear regression model
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(X_train_scaled,y_train)
y_pred=linreg.predict(X_test_scaled)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
mae,r2
```

```python
plt.scatter(y_test,y_pred)
plt.show()
```

```python
#lasso regression
from sklearn.linear_model import Lasso
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
```

```python
y_pred=lasso.predict(X_test_scaled)
y_pred
```

```python
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
mae,r2
```

```python
plt.scatter(y_test,y_pred)
plt.show()
```

```python
#lasso regression with cross validation
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train_scaled,y_train)
y_pred=lassocv.predict(X_test_scaled)
y_pred
```

```python
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
mae,r2
```

```python

plt.scatter(y_test,y_pred)
plt.show()
```
